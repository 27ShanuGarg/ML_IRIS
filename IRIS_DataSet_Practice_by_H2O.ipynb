```python
target_map = {0:'A', 1:'B', 2:'C'}
# df['target']
df['target'] = df['target'].map(target_map)
df['target']
```

```python
from h2o.automl import H2OAutoML
# cols = df.columns.tolist()
# df = df[df.target !=2][cols]
# df

h2o_df = h2o.H2OFrame(df)
# h2o_df = h2o.import_file('/content/Iris.csv')
h2o_df
```

```python
h2o_df.types
```

```python
h2o_df.describe(chunk_summary=True)
```

```python
churn_df = h2o_df
churn_train,churn_test,churn_valid = churn_df.split_frame(ratios=[.7, .15])
# train, test = h2o_df.split_frame(ratios=[0.8], seed = 1)
churn_train.shape,churn_test.shape,churn_valid.shape
```

```python
y = "target"
x = churn_df.columns
x.remove(y)
# x.remove('Id')
x, y
```

```python
aml = H2OAutoML(max_models = 10, exclude_algos = ["StackedEnsemble", "DeepLearning"], verbosity="info", nfolds=0,distribution="multinomial",seed=1234)
```

```python
aml.train(x = x, y = y, training_frame = churn_train, validation_frame=churn_valid)
# aml.train(training_frame = train, y = y)
```

```python
# Get leaderboard with all possible columns
lb = h2o.automl.get_leaderboard(aml, extra_columns = "ALL")
lb
```

```python
churn_pred=aml.leader.predict(churn_test)
churn_pred.head()
```

```python
aml.leader.model_performance(churn_test)
```

```python
model_ids = list(aml.leaderboard['model_id'].as_data_frame().iloc[:,0])
model_ids
```

```python
h2o.get_model([mid for mid in model_ids if "XGBoost" in mid][0])
```

```python
out_glm = h2o.get_model([mid for mid in model_ids if "GLM" in mid][0])
out_glm
```
